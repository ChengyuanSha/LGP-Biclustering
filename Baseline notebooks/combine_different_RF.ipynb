```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set()
```

```python
df0 = pd.read_csv('rf0.csv', header=None, index_col=0)
df1 = pd.read_csv('rf1.csv', header=None, index_col=0)
df2 = pd.read_csv('rf2.csv', header=None, index_col=0)
df3 = pd.read_csv('rf3.csv', header=None, index_col=0)
df4 = pd.read_csv('rf4.csv', header=None, index_col=0)
```

```python
pd_group = pd.concat([df0, df1, df2, df3, df4])
group_info = pd_group.groupby(pd_group.index)
df_means = group_info.mean().sort_values(by=1, ascending=False)
```

```python
df_means = df_means.head(20)
```

```python
ax = sns.barplot(data = df_means, x = 1, y = df_means.index, palette="Blues_d")
ax.set(xlabel='Importance', ylabel='Features')
plt.tight_layout()
plt.savefig('f_rank_rf_avg.tif', dpi=300)
plt.show()
```
